```python
import pandas as pd
from sentence_transformers import SentenceTransformer
import warnings
warnings.filterwarnings('ignore')
```

```python

df2=pd.read_csv("C:/Users/ashis/OneDrive/Desktop/tweeter/Twitter_Data.csv")
```

```python
df2=df2.sample(1000)
```

```python
df2
```

```python
df2.dropna(inplace=True)
```

```python
df2['clean_text']=df2['clean_text'].str.strip()
```

```python
list(df2['clean_text'].values)
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
df2['embeddings']=df2['clean_text'].apply(lambda x:model.encode(x,convert_to_numpy=True))
```

```python
df2
```

```python
from sklearn.model_selection import train_test_split
df_train,df_test=train_test_split(df2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
clf=KNeighborsClassifier(5)
```

```python
embeddings=model.encode(df_train.clean_text.tolist(),convert_to_numpy=True)
```

```python
clf.fit(embeddings,df_train.category.tolist())
```

```python
embeddings2=model.encode(df_test.clean_text.tolist(),convert_to_numpy=True)
```

```python
clf.score(embeddings2,df_test.category.tolist())
```

```python
test=('ashish is a good person')
enc=model.encode(test)
review=clf.predict([enc])
```

```python
if review==0:
    print("Neutral comment")
elif review==1:
    print("Postive comment")
elif review==-1:
    print("Negative comment")
```
